```python
features = ['latitude', 'longitude',
       'minimum_nights', 'number_of_reviews', 'reviews_per_month',
       'calculated_host_listings_count', 'availability_365', 'neighbourhood_group', 'room_type']
    
for feature in features:
    
    df_train_model = df_train.copy()
    df_val_model = df_val.copy()
    
    del df_train_model[feature]
    del df_val_model[feature]
    
    combined_features = features.copy()
    combined_features.remove(feature)
    
    train_dicts = df_train_model[combined_features].to_dict(orient='records')
    val_dicts = df_val[combined_features].to_dict(orient='records')

    dv = DictVectorizer()
    dv.fit(train_dicts)

    X_train = dv.transform(train_dicts)
    X_val = dv.transform(val_dicts)

    model = LogisticRegression(solver='liblinear', C=1.0, random_state=42)
    model.fit(X_train, y_train)

    y_predict = model.predict(X_val)

    modal_accuracy = (y_val == y_predict).mean() * 100
    distance = original_accuracy - modal_accuracy
    
    print(feature + ': ' + str(distance))
```

```python
from sklearn.model_selection import train_test_split

df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df, test_size=0.33, random_state=42)

y_train = df_train.price.values
y_val = df_val.price.values
y_test = df_test.price.values

del df_train['price']
del df_val['price']
del df_test['price']
```

```python
df_train = df_train.fillna(0)
df_val = df_val.fillna(0)
df_test = df_test.fillna(0)
```

```python
def RMSE(y_pred, y_train):
    
    m = len(y_train)
    
    diff = y_pred - y_train
    squared_error = np.power(diff, 2)
    squared_error_sum = squared_error.sum()
    mean_square_error = (1/m) * squared_error_sum
    
    return np.sqrt(mean_square_error)
```

```python
from sklearn.linear_model import Ridge

alphas = [0, 0.01, 0.1, 1, 10]

for i in alphas:

    train_dicts = df_train[categorical + numerical].to_dict(orient='records')
    val_dicts = df_val[categorical + numerical].to_dict(orient='records')

    dv = DictVectorizer()
    dv.fit(train_dicts)

    X_train = dv.transform(train_dicts)
    X_val = dv.transform(val_dicts)

    model = Ridge(alpha=i)
    model.fit(X_train, y_train)

    y_predict = model.predict(X_val)

    rmse = RMSE(y_predict, y_val).round(3)
    
    print('Alpha ' + str(i) + ': ' + 'RMSE = ' + str(rmse))
```

Alpha 0 is the answer!
