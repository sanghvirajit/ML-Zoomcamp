```python
df = data[['neighbourhood_group', 'room_type', 'latitude', 'longitude', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365', 'price']]
df.head()
```

```python
df = df.fillna(0)
df.isnull().sum()
```

```python
df['price'] = np.log1p(df.price.values)
df.head()
```

```python
from sklearn.model_selection import train_test_split

df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=1)

df_train_full = df_train_full.reset_index(drop=True)
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.price.values
y_val = df_val.price.values
y_test = df_test.price.values

del df_train['price']
del df_val['price']
del df_test['price']
```

```python
from sklearn.feature_extraction import DictVectorizer

train_dicts = df_train.to_dict(orient='records')
val_dicts = df_val.to_dict(orient='records')

dv = DictVectorizer(sparse=False)

X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)
```

# Question 2

```python
def RMSE(y_pred, y_train):
    
    m = len(y_train)
    
    diff = y_pred - y_train
    squared_error = np.power(diff, 2)
    squared_error_sum = squared_error.sum()
    mean_square_error = (1/m) * squared_error_sum
    
    return np.sqrt(mean_square_error)
```

```python
from sklearn.ensemble import RandomForestRegressor

estimators = 10

rf = RandomForestRegressor(n_estimators=estimators,
                            random_state=1)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_val)
RMSE_ = RMSE(y_pred, y_val)

print("RMSE: " + str(RMSE_))
```

**0.4615** is the answer

# Question 3

```python
scores = []

for n in range(10, 201, 10):
    rf = RandomForestRegressor(n_estimators=n, random_state=1)
    rf.fit(X_train, y_train)

    y_pred = rf.predict(X_val)
    RMSE_ = RMSE(y_pred, y_val)
    
    scores.append((n, RMSE_))
```

```python
df_scores = pd.DataFrame(scores, columns=['n_estimators', 'RMSE'])

plt.plot(df_scores.n_estimators, df_scores.RMSE)
```

**120** is the answer

# Question 4

```python
scores = []

for d in [10, 15, 20, 25]:
    for n in range(10, 201, 10):
        rf = RandomForestRegressor(n_estimators=n,
                                    max_depth=d,
                                    random_state=1
                                    n_jobs=-1)
        rf.fit(X_train, y_train)

        y_pred = rf.predict(X_val)
        RMSE_ = RMSE(y_pred, y_val)
    
        scores.append((d, n, RMSE_))
```

```python
columns = ['max_depth', 'n_estimators', 'RMSE']
df_scores = pd.DataFrame(scores, columns=columns)
```

```python
for d in [10, 15, 20, 25]:
    df_subset = df_scores[df_scores.max_depth == d]
    
    plt.plot(df_subset.n_estimators, df_subset.RMSE,
             label='max_depth=%d' % d)

plt.legend()
```

**15** is the best depth

**answer will not be different if we change the seed for the model**

```python
def parse_xgb_output(output):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))
    
    columns = ['num_iter', 'train_rmse', 'val_rmse']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
```
